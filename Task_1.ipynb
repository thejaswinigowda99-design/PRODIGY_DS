```python
# Load the uploaded CSV and quickly inspect to pick a variable to visualize, then plot a distribution
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the file (utf-8 per metadata)
df_population = pd.read_csv('C:/Users/hp/Downloads/merged_population_data.csv', encoding='utf-8')

# Show head to confirm structure
print(df_population.head())

# Choose a continuous variable: year 2020 population values (wide format per preview)
# Coerce to numeric and drop NaNs for plotting
pop_2020 = pd.to_numeric(df_population.get('2020'), errors='coerce').dropna()

# Basic histogram of 2020 population by country/economy
plt.figure(figsize=(9,5))
sns.histplot(pop_2020, bins=40, kde=True, color='steelblue')
plt.title('Distribution of Total Population in 2020')
plt.xlabel('Population (people)')
plt.ylabel('Count of entities')
plt.tight_layout()
plt.show()

print('Loaded merged_population_data.csv, displayed head, and plotted a histogram for 2020 population.')
```
